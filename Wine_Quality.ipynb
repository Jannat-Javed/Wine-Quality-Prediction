## Input Split

```python
X = df.drop(columns=['type', 'quality'])
y = df['quality']
```

```python
X
```

```python
y
```

## Class Imbalance

```python
y.value_counts()
```

```python
from imblearn.over_sampling import SMOTE
oversample = SMOTE(k_neighbors=4)
# transform the dataset
X, y = oversample.fit_resample(X, y)
```

```python
y.value_counts()
```

## Train Test Split

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
X_train
```

## Scalling

```python
from sklearn.preprocessing import StandardScaler

# Initialize the scaler
scaler = StandardScaler()

# Fit the scaler on the training data and transform both X_train and X_test
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

## Model Training

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier,GradientBoostingClassifier,ExtraTreesClassifier
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score

def classify(model, X_train_scaled, X_test_scaled, y_train, y_test):
    # Train the model
    model.fit(X_train_scaled, y_train)
    
    # Predict and evaluate the model
    y_pred = model.predict(X_test_scaled)
    print("Accuracy:", model.score(X_test_scaled, y_test) * 100)
    
    # Cross-validation
    score = cross_val_score(model, X_train_scaled, y_train, cv=5)
    print("CV Score:", np.mean(score) * 100)
```

## Decision Tree

```python
decision_tree = DecisionTreeClassifier()
classify(decision_tree, X_train_scaled, X_test_scaled, y_train, y_test)
```

## Random Forest

```python
random_forest = RandomForestClassifier()
classify(random_forest, X_train_scaled, X_test_scaled, y_train, y_test) 
```

## Extra Trees

```python
extra_trees = ExtraTreesClassifier()
classify(extra_trees, X_train_scaled, X_test_scaled, y_train, y_test)
```

## KNN

```python
knn = KNeighborsClassifier()
classify(knn, X_train_scaled, X_test_scaled, y_train, y_test)
```

## Gradient Boosting

```python
gradient_boosting = GradientBoostingClassifier()
classify(gradient_boosting, X_train_scaled, X_test_scaled, y_train, y_test)
```

## Voting Classifier

```python
from sklearn.ensemble import VotingClassifier
rf = RandomForestClassifier(n_estimators=350,random_state=3,max_samples=0.5,max_features=0.75,max_depth=15)
gbdt = GradientBoostingClassifier(n_estimators=100,max_features=0.5)
et = ExtraTreesClassifier(n_estimators=100,random_state=3, bootstrap=True,max_samples=0.5,max_features=0.75,max_depth=15)

voting_clf = VotingClassifier([('rf', rf), ('gbdt', gbdt), ('et',et)])

# Train the VotingClassifier
voting_clf.fit(X_train_scaled, y_train)

# Make predictions
y_pred = voting_clf.predict(X_test_scaled)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy * 100)

# Optionally, you can perform cross-validation
cross_val_scores = cross_val_score(voting_clf, X_train_scaled, y_train, cv=5)
print("CV Score:", np.mean(cross_val_scores) * 100)
```
